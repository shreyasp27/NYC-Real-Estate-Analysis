```python
def findpoint(df):

   def closest_point(point, points):
       """ Find closest point from a list of points. """
       return points[cdist([point], points).argmin()]

   def match_value(df, col1, x, col2):
       """ Match value x from col1 row to value in col2. """
       return df[df[col1] == x][col2].values[0]
   
   df['point'] = [(x) for x in zip(df['GROSS SQUARE FEET'],df['Cluster'])]
   cd2['point'] = [(x) for x in zip(cd2['GROSS SQUARE FEET'],cd2['Cluster'])]
   cd2['closest_GSF'] = [closest_point(x, list(df['point'])) for x in cd2['point']]
   cd2['Cluster'] = [match_value(df, 'point', x, 'Cluster') for x in cd2['closest_GSF']]
   cd2['RESIDENTIAL UNITS'] = [match_value(df, 'point', x, 'RESIDENTIAL UNITS') for x in cd2['closest_GSF']]
   cd2['COMMERCIAL UNITS'] = [match_value(df, 'point', x, 'COMMERCIAL UNITS') for x in cd2['closest_GSF']]
   cd2['TOTAL UNITS'] = [match_value(df, 'point', x, 'TOTAL UNITS') for x in cd2['closest_GSF']]
   cd2['LAND SQUARE FEET'] = [match_value(df, 'point', x, 'LAND SQUARE FEET') for x in cd2['closest_GSF']]
   cd2['BUILDING AGE'] = [match_value(df, 'point', x, 'BUILDING AGE') for x in cd2['closest_GSF']]
   cd2['NEIGHBORHOOD'] = [match_value(df, 'point', x, 'NEIGHBORHOOD') for x in cd2['closest_GSF']]
   cd2['ADDRESS'] = [match_value(df, 'point', x, 'ADDRESS') for x in cd2['closest_GSF']]
   cd2['SALE PRICE'] = [match_value(df, 'point', x, 'SALE PRICE') for x in cd2['closest_GSF']]
```
